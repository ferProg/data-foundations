# 03_distributions_skewness_and_outliers

This notebook explores distribution shape, skewness, and outliers in quantitative data.

Objectives:
- Visualize distribution shape
- Identify skewness
- Detect outliers
- Discuss interpretation limits


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/example_distribution.csv')
df.head()
```

```python
df.describe()
```

```python
plt.boxplot(df['value'], vert=False)
plt.title('Boxplot of Values')
plt.xlabel('Value')
plt.show()
```

## Interpretation

The histogram shows a right-skewed distribution, with most values concentrated in the lower range.
The boxplot makes extreme values clearly visible as outliers.
Skewness affects the mean more than the median, pulling it toward higher values.
Outliers may represent rare but valid observations or data quality issues.
Without context, it is not possible to decide whether outliers should be removed.
Visualizations help detect structure but do not explain underlying causes.
