# Interpreting a Linear Regression Model

Fitting a regression model is only the first step.
Understanding its coefficients and residuals is essential to evaluate
whether the model is meaningful and appropriate.


In the previous notebook, a simple linear regression was fitted to describe
the relationship between two quantitative variables. Here, we focus on
interpreting the fitted model rather than on improving its performance.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("../data/example_relationships.csv")

X = df[['x']]
y = df['y']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
```

## Interpreting the Coefficient

The slope of the regression line represents the expected average change
in the response variable for a one-unit increase in the predictor.

This interpretation is valid only within the range of observed data
and does not imply a causal relationship.


```python
slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```

## Interpreting the Intercept

The intercept represents the expected value of the response when the
predictor is equal to zero. In many real-world cases, this value has
little practical meaning and should be interpreted cautiously.


## Residual Analysis

Residuals are the differences between observed values and model predictions.
Analyzing residuals helps assess whether model assumptions are reasonable.


```python
residuals = y - y_pred

plt.scatter(y_pred, residuals)
plt.axhline(0)
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
plt.show()
```

If residuals are randomly scattered around zero, the linear model may be
a reasonable approximation. Systematic patterns, increasing spread, or
clusters in the residuals can indicate violations of model assumptions
or missing structure in the data.


## Limits of Interpretation

Even when coefficients and residuals appear well-behaved, a simple linear
model remains a simplification. Interpretation should remain grounded in
the context of the data and its limitations.
