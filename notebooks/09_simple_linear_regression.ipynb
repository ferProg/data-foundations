# From Exploration to Modeling: Simple Linear Regression


After exploratory analysis, models can be used to formalize relationships
between variables. A simple linear regression is often the first step in
this direction.


## Model Assumptions

A simple linear regression assumes:

- A linear relationship between variables
- Independence of observations
- Constant variance of errors
- No strong outliers dominating the fit


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('../data/example_relationships.csv')
X = df[['x']]
y = df['y']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("X variable")
plt.ylabel("Y variable")
plt.show()
```

## Interpretation

The fitted line represents the average change in the response variable
associated with a one-unit change in the predictor. This does not imply
a causal relationship and should be interpreted cautiously.


## Limitations

- The model does not capture non-linear patterns
- Correlation does not imply causation
- Predictions outside the observed range are unreliable
