# 01_reading_tables_and_categorical_data

This notebook demonstrates how to load a simple CSV and explore categorical data.

Objectives:
- Load a CSV file
- Create a frequency table
- Create a bar plot
- Document observations about the data


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load simple dataset
df = pd.read_csv('../data/example_categorical.csv')

# Show first rows
df.head()
```

```python
# Frequency table of the categorical variable
frequency_table = df['category'].value_counts()
frequency_table
```

```python
# Bar plot
frequency_table.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.title('Category Frequencies')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

## Observations

- What you can see: the most and least frequent categories, general distribution of the data.
- What you cannot conclude: no causal relationships or trends outside of this dataset can be inferred.
