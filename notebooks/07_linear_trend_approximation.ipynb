# 07_linear_trend_approximation

This notebook introduces a simple linear trend as a way to approximate the relationship between two quantitative variables.

Objectives:
- Approximate a linear trend
- Visualize observed data and trend together
- Discuss the meaning and limitations of the approximation


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/example_relationships.csv')
df.head()
```

```python
plt.scatter(df['x'], df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X vs Y')
plt.show()
```

```python
# Compute linear trend using numpy
coefficients = np.polyfit(df['x'], df['y'], 1)
slope, intercept = coefficients

# Generate trend line
x_values = np.array(df['x'])
y_trend = slope * x_values + intercept

slope, intercept
```

```python
plt.scatter(df['x'], df['y'], label='Observed data')
plt.plot(x_values, y_trend, label='Linear trend')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Trend Approximation')
plt.legend()
plt.show()
```

## Interpretation

The linear trend summarizes the overall direction of the relationship between X and Y.
It provides a simplified approximation of how Y changes as X increases.
Individual points deviate from the trend, indicating variability not explained by the model.
The trend does not capture all patterns present in the data.
This approximation does not imply causation or predictive accuracy.
Linear trends are useful for description, not for definitive explanation.
