# 04_comparing_distributions_across_groups

This notebook compares quantitative distributions across different groups.

Objectives:
- Compare distributions between groups
- Visualize differences in central tendency and spread
- Discuss interpretation limits


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/example_groups.csv')
df.head()
```

```python
df.groupby('group')['value'].describe()
```

```python
plt.figure(figsize=(6,4))
df.boxplot(column='value', by='group')
plt.title('Value Distribution by Group')
plt.suptitle('')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()
```

## Interpretation

Group B shows higher values on average compared to Group A.
The boxplots suggest differences in both central tendency and spread.
However, visual separation does not imply causality.
The sample sizes are small, which limits generalization.
Without statistical testing, differences should be considered descriptive only.
Context is required to understand why groups differ.
