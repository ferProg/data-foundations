# 06_relationships_between_quantitative_variables

This notebook explores the relationship between two quantitative variables using scatter plots.

Objectives:
- Visualize relationships between variables
- Identify trends and variability
- Discuss interpretation limits


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/example_relationships.csv')
df.head()
```

```python
plt.scatter(df['x'], df['y'])
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Interpretation

The scatter plot shows a positive relationship between X and Y.
As X increases, Y tends to increase as well.
The relationship is not perfectly linear, indicating variability.
The plot suggests association but does not imply causation.
Other variables not shown here may influence the relationship.
Visual patterns help guide further analysis but are not conclusive.
