# Student Performance — 04 Conclusions and Limits

This notebook summarizes key findings from the exploratory analysis and modeling,
and documents limitations and next steps.

## What we did

- Loaded and cleaned the Student Performance dataset (Math course).
- Performed exploratory analysis focused on the final grade (G3).
- Trained simple linear regression models under two scenarios:
  - **Late-stage prediction**: including G1 and G2
  - **Early-stage prediction**: excluding G1 and G2 to reduce leakage
- Compared train/test performance to assess generalization.

## Main findings

- Final grades (G3) show meaningful variation across students and are influenced by multiple factors.
- Including earlier grades (G1, G2) substantially improves predictive performance, as expected.
- Without G1 and G2, performance drops, suggesting that background/study variables alone explain a more limited portion of G3.

## Key insight: G1/G2 and information leakage

G1 and G2 are strong predictors of G3 because they reflect earlier academic outcomes.
Whether they should be included depends on the use-case:

- If predicting G3 near the end of the year (when G1/G2 are known), including them is realistic.
- If predicting early (before grading periods), including them would be unrealistic and may constitute information leakage.

This project explicitly evaluates both scenarios to keep interpretation honest.

## Limitations

- This analysis is observational: it does not establish causal relationships.
- Models are simple and use a single train/test split (results may vary with different splits).
- Only numeric features were used in the first modeling pass; categorical variables may carry additional signal.
- The dataset represents a specific context (schools, time period) and may not generalize to other populations.

## Next steps

- Add categorical variables via one-hot encoding to improve early-stage modeling.
- Try regularized linear models (Ridge/Lasso) to handle correlated predictors.
- Use cross-validation for more stable performance estimates.
- Explore model diagnostics (residual patterns, influential points).

## Final note

The main value of this project is demonstrating a disciplined workflow:
exploration → cautious modeling → evaluation → honest conclusions.
Performance metrics are treated as summaries, not as proof of truth.
