# 02_describing_quantitative_data

This notebook explores basic descriptive statistics and visualizations for quantitative data.

Objectives:
- Visualize distributions using histograms and boxplots
- Compute mean and median
- Compare how different visualizations affect interpretation


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/example_quantitative.csv')
df.head()
```

```python
mean_value = df['value'].mean()
median_value = df['value'].median()

mean_value, median_value
```

```python
plt.hist(df['value'], bins=6)
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(df['value'], vert=False)
plt.title('Boxplot of Values')
plt.xlabel('Value')
plt.show()
```

## Interpretation across visualizations

The histogram shows the overall shape of the distribution and suggests a slight right skew.
The mean is higher than the median, which is consistent with this skewness.
The boxplot highlights the spread and makes potential outliers easier to detect.
While the histogram emphasizes frequency patterns, the boxplot focuses on central tendency and dispersion.
Depending on the visualization, the same data can appear more or less symmetric.
This shows that interpretation depends strongly on how data is represented.
