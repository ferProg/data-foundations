# 05_variability_iqr_and_standard_deviation

This notebook explores different measures of variability in quantitative data.

Objectives:
- Compute range, IQR, and standard deviation
- Understand what each metric captures
- Compare robustness to extreme values


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/example_variability.csv')
df.head()
```

```python
df.describe()
```

```python
value_range = df['value'].max() - df['value'].min()

q1 = df['value'].quantile(0.25)
q3 = df['value'].quantile(0.75)
iqr = q3 - q1

value_range, iqr
```

```python
# standard deviation
std_dev = df['value'].std()
std_dev
```

```python
plt.boxplot(df['value'], vert=False)
plt.title('Boxplot of Values')
plt.xlabel('Value')
plt.show()
```

## Interpretation

The range captures the total spread but is highly sensitive to extreme values.
The interquartile range focuses on the middle 50% of the data and is more robust.
Standard deviation measures average dispersion around the mean.
Different measures highlight different aspects of variability.
A dataset can have a small IQR but a large range due to outliers.
Choosing a variability measure depends on data shape and analysis goals.
